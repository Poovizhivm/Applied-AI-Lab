```python
# Inputs
x1, x2, x3, x4 = 1, 1, 0, 1
X = [x1, x2, x3, x4]
# Output
y = [1,0]
# Learning rate
lr = 0.84
# Weight
w15, w16 = 0.3, 0.1
w25, w26 = -0.2, 0.4
w35, w36 = 0.2, -0.3
w45, w46 = 0.1, 0.4
w57, w58 = -0.3, 0.2
w67, w68 = 0.3, 0.1
w79, w89 = -0.4, 0.3
w710, w810 = 0.2, -0.1

# bias
b5, b6, b7, b8, b9, b10 = 0.2, 0.1, -0.3, 0.2, 0.5, 0.3

max_epochs = 10
```

```python
import math 
def sigmoid(x): 
    return 1 / (1 + math.exp(-x)) 
def softmax(z): 
    exp_vals = [math.exp(i) for i in z] 
    sum_exp = sum(exp_vals) 
    return [val / sum_exp for val in exp_vals]
```

```python
for epoch in range(max_epochs):
    # Hidden Layer 1
    net5 = x1*w15 + x2*w25 + x3*w35 + x4*w45 + b5
    net6 = x1*w16 + x2*w26 + x3*w36 + x4*w46 + b6
    x5 = sigmoid(net5)
    x6 = sigmoid(net6)
    
    # Hidden Layer 2
    net7 = x5*w57 + x6*w67 + b7
    net8 = x5*w58 + x6*w68 + b8
    x7 = sigmoid(net7)
    x8 = sigmoid(net8)
    
    # Output Layer (2 neurons)
    net9  = x7*w79  + x8*w89  + b9
    net10 = x7*w710 + x8*w810 + b10
    outputs = softmax([net9, net10])   
    
    error = [yt - o for yt, o in zip(y, outputs)]
    print(f"Epoch {epoch}: Output = {[f'{o:.3f}' for o in outputs]}, Error = {[f'{e:.3f}' for e in error]}")

    # BACKPROPAGATION
    delta9  = error[0]   
    delta10 = error[1]   

    # Hidden layer 2 deltas
    delta7 = (delta9 * w79 + delta10 * w710) * x7 * (1 - x7)
    delta8 = (delta9 * w89 + delta10 * w810) * x8 * (1 - x8)

    # Hidden layer 1 deltas
    delta5 = (delta7 * w57 + delta8 * w58) * x5 * (1 - x5)
    delta6 = (delta7 * w67 + delta8 * w68) * x6 * (1 - x6)

    # WEIGHT & BIAS UPDATES 
    # Output layer
    w79  += lr * delta9  * x7
    w89  += lr * delta9  * x8
    b9   += lr * delta9

    w710 += lr * delta10 * x7
    w810 += lr * delta10 * x8
    b10  += lr * delta10

    # Hidden layer 2
    w57 += lr * delta7 * x5
    w67 += lr * delta7 * x6
    b7  += lr * delta7

    w58 += lr * delta8 * x5
    w68 += lr * delta8 * x6
    b8  += lr * delta8

    # Hidden layer 1
    w15 += lr * delta5 * x1
    w25 += lr * delta5 * x2
    w35 += lr * delta5 * x3
    w45 += lr * delta5 * x4
    b5  += lr * delta5

    w16 += lr * delta6 * x1
    w26 += lr * delta6 * x2
    w36 += lr * delta6 * x3
    w46 += lr * delta6 * x4
    b6  += lr * delta6

print("\nFinal Output:", [f"{o:.3f}" for o in outputs])
```

```python
import numpy as np

# Step activation function
def step_function(net):
    return 1 if net >= 0 else 0
    
# Perceptron training function
def train_perceptron(X, T, w, b, lr):
    epoch = 0
    while True:
        error_count = 0
        print(f"\nEpoch {epoch}")
        for i in range(len(X)):
            net = np.dot(X[i], w) + b
            y = step_function(net)
            error = T[i] - y
            print(f"Input: {X[i]}, Net: {net:.2f}, Output: {y}, Target: {T[i]}")
            if error != 0:
                w = w + lr * error * X[i]
                b = b + lr * error
                error_count += 1
        if error_count == 0:
            break
        epoch += 1
    return w, b
```

```python
# Input patterns
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
learning_rate = 0.1
```

```python
#OR GATE
print("OR Gate")
T_or = np.array([0, 1, 1, 1])
w_or = np.array([0.2, 0.2])
b_or = -0.1
final_w_or, final_b_or = train_perceptron(X, T_or, w_or, b_or, learning_rate)
print("\nFinal OR Gate Weights:", final_w_or)
print("Final OR Gate Bias:", final_b_or)
```

```python
#AND GATE 
print("AND Gate")
T_and = np.array([0, 0, 0, 1])
w_and = np.array([0.2, 0.2])
b_and = -0.3
final_w_and, final_b_and = train_perceptron(X, T_and, w_and, b_and, learning_rate)
print("\nFinal AND Gate Weights:", final_w_and)
print("Final AND Gate Bias:", final_b_and)
```
